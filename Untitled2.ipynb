```python
#Draw the line chart of fresh tomatoes and fresh potatoes consumption from year 1970 to 2010.
```

```python

plt.plot(data1['Year'], data1['Fresh_tomatoes'], linestyle="solid", marker='o', color="red", label = 'Fresh tomatoes')
plt.plot(data2['Year'], data2['Fresh_potatoes'], linestyle="solid", marker='o', color="green", label = 'Fresh potatoes')
plt.xlabel("Year")
plt.ylabel("Consumption")
plt.legend()
plt.title("Consumption Analysis")
plt.show()
```

```python
#Draw line chart of fresh and processed tomatoes from year 1980 to 1999.
```

```python
plt.plot(data1['Year'], data1['Fresh_tomatoes'], linestyle="solid", marker='o', color="red", label = 'Fresh tomatoes')
plt.plot(data1['Year'], data1['Processed_tomatoes'], linestyle="solid", marker='o', color="green", label = 'Processed tomatoes')
plt.xlim(1980,1999)
plt.xlabel("Year")
plt.ylabel("Consumption")
plt.legend()
plt.title("Consumption Analysis")
plt.show()
```
